```python
# Chart - 1 visualization code
# Barplot on channel_name

md1 = md[md.channel_name=='Outcall'].groupby('CSAT Score')['CSAT Score'].count()
md2 = md[md.channel_name=='Inbound'].groupby('CSAT Score')['CSAT Score'].count()
md3 = md[md.channel_name=='Email'].groupby('CSAT Score')['CSAT Score'].count()

md11 = pd.DataFrame(md1)


md11.rename(columns={'CSAT Score': 'count'}, inplace=True)
md11.reset_index(inplace=True)

plt.figure(figsize=(12,10))

sns.barplot(
    data=md11,
    x='CSAT Score',
    y='count',
    hue='CSAT Score',
    palette='Set2'   \
)

plt.xlabel("CSAT SCORE")
plt.ylabel("OUTCALL Channel Type")
plt.title("Count of Outcall vs CSAT Score")

plt.show()
```

1. Why did you pick the specific chart?

We selected a bar plot because it helps in comparing the count of CSAT scores for the Outcall channel easily. Bar charts give a clear visual comparison between different score values and make the data easier to understand.

2. What are the insights found from the chart?

From the bar plot, the following insights were observed:
	•	CSAT Score 5 has the highest count for the Outcall channel, which is above 10,000.
	•	CSAT Score 2 has the lowest count, which is less than 500.
	•	CSAT Score 1 and 4 have almost similar counts, around 2000.
	•	Most customers have given higher scores like 5 and 4, while fewer customers have given scores 2 and 3.

3. Will the gained insights help create a positive business impact?

Yes, the insights can help in understanding both positive and negative impacts.

On the positive side, a large number of customers have given CSAT scores of 4 and 5, which shows that many customers are satisfied with the service provided.

However, there is also a negative aspect, as the count of CSAT score 1 is still noticeable, which means some customers are not satisfied with the support they received. The company should focus on improving customer service to reduce low satisfaction scores.

```python
# Chart - 10 visualization code

# Convert groupby Series to DataFrame and reset index
df = md.groupby(["Product_category", "CSAT Score"]).size().unstack(fill_value=0)

# Convert values into percentage
dfu = df.div(df.sum(axis=1), axis=0) * 100

# Plot stacked bar chart
fig, ax = plt.subplots(figsize=(9,7))

dfu.plot(
    kind="barh",
    stacked=True,
    colormap="viridis",   
    ax=ax
)

# Labels and title
plt.xlabel("Percentage of CSAT Score")
plt.ylabel("Product Category")
plt.title("Product Category vs CSAT Score")

plt.legend(title="CSAT Score", bbox_to_anchor=(1.05,1), loc="upper left")

plt.grid(axis="x")

plt.show()
```

1. Why did you pick the specific chart?

A stacked horizontal bar chart was chosen because it helps to compare CSAT scores across different product categories in a single chart. It clearly shows how customer satisfaction is distributed for each category.

2. What insights were found from the chart?
	•	Some product categories have higher CSAT scores (4 and 5) compared to others.
	•	A few categories also show lower CSAT scores like 1 and 2, indicating dissatisfaction.
	•	This helps identify which product categories are performing well and which need improvement.

3. Will the insights help create a positive business impact?
